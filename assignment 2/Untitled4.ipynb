```python
import pandas as pd
import matplotlib.pyplot as plt  
import numpy as np
```

```python
# import data
df=pd.read_csv("..\..\Downloads\Public_Secondary_Repeaters_by_Class.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df1=df.groupby(['county','boysform1','boysform2']).sum()

df1.reset_index(inplace=True)
df1.head()
```

```python
df['sum_boys']=(df['boysform1'] + df['boysform2'] + df['boysform3'] + df['boysform4'])
df
```

```python
df['sum_girls']=(df['girlsform1'] + df['girlsform2'] + df['girlsform3'] + df['girlsform4'])
df
```

```python
df['sum_county']=df['sum_boys']+df['sum_girls']
df
```

```python
df1=df.groupby(['county','sum_boys', 'sum_girls','sum_county']).sum()

df1.reset_index(inplace=True)
df1.head()
```

```python
df1.plot(x='county',y=['sum_boys','sum_girls'],kind='bar',figsize=(15,8))
plt.title("sum of Public seconday school repeaters per county: Boys vs Girls")
plt.xlabel("County name")
plt.ylabel("Number of repeating students")
plt.show()
```

```python
df1.plot(x='county',y=['sum_county'],kind='bar',figsize=(15,8), color='red')
plt.title("Distribution of Public seconday school repeaters per county :boys & girls")
plt.xlabel("County name")
plt.ylabel("Number of repeating students")
plt.show()
```

```python
df1.plot(x='county',y=['boysform1','girlsform1'],kind='bar',figsize=(15,8))
plt.title("Form 1 distribution of Public seconday school repeaters per county: boys vs girls")
plt.xlabel("County name")
plt.ylabel("Number of repeating students")
plt.show()
```
