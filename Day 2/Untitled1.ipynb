# House Hunting

## scenario 1

```python
#House Hunting
import math # import the functions
current_savings = 0 #your current savings
investment_r = 0.04 #investment growth rate
total_cost = int(input("What is the cost of your dream home?")) #the total cost of your dream home
annual_salary = float(input("What is your annual salary?")) #your annual salary
monthly_savings = int(input("What is your monthly savings?")) #monthly savings
monthly_salary = annual_salary/12
monthly_rate = investment_r / 12
monthly_interest = monthly_rate * monthly_salary
print("The monthly salary is: %s" % monthly_salary)
portion_saved = monthly_savings/annual_salary
print("The portion saved as a fraction of annual salary is: %s" % portion_saved)
yearly_savings = monthly_savings * 12
print("The yearly savings is: %s" % yearly_savings)
portion_down_payment = 0.25 * total_cost #house down payment is 25% of the total house cost
print("The deposit needed is: %s" % portion_down_payment)
month = 0
while current_savings < portion_down_payment:
    month += 1
    monthly_interest = current_savings * monthly_rate
    current_savings += monthly_savings + monthly_interest
print("It will take you {} months to save enough money for the downpayment" . format(month))
```
